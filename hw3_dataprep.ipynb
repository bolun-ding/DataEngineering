```python
#Go through all the csv in the folder and slice them by the start and end date
for f in os.listdir():
    if 'historical_fx_rates' in f and not f.endswith('.zip'):
        print(f)
        w = 0
        for file in os.listdir(path=f):
            if file.endswith('.csv'):
                print(file)
                df = pd.read_csv(f'{f}/{file}')
                df = df[(df['datetime'] >= start_date) & (df['datetime'] <= end_date)]
                df['hour']= pd.to_datetime(df['datetime']).dt.floor('H')
                #replace the NaN with average value of the column
                check_na_col = ['v', 'vw', 'o', 'c', 'h', 'l', 't', 'n',]
                for col in check_na_col:
                    if df[col].isnull().sum() > 0:
                        df[col] = df[col].fillna(df[col].mean())
                print('Data loaded')
                #calculate the transaction_vmap
                df['transaction_vmap'] = df['vw'] * df['v']

                #calculate VWAP and Liquidity by hour
                hourly_vmap = df.groupby('hour').apply(lambda x:{'VMAP': x['transaction_vmap'].sum() / x['v'].sum(),
                                                       'liquidity': x['n'].sum()}).reset_index().rename(columns={0: 'data'})
                
                hourly_vmap = pd.concat([hourly_vmap.drop(['data'], axis=1), hourly_vmap['data'].apply(pd.Series)], axis=1)
                print('VWAP and Liquidity calculated')
                #create collection name according to the file name
                collection_name = file.split('.')[0]
                collection = db[collection_name]

                collection.delete_many({})
                records = hourly_vmap.to_dict('records')
                collection.insert_many(records)
                print(str(w) + ' inserted into ' + collection_name)
                w += 1
print('All data inserted into MongoDB')
```

```python
#Export the data into CSV file, which can be directly loaded into the pycaret
from tqdm import tqdm
#initialize the results as DataFrame
results_all = pd.DataFrame()
result_all = {'pair':[], 'Timestamp':[], 'VWAP':[],
            'Liquidity':[], 'volatility':[], 'max':[], 'min':[], 'FD':[]}
for f in os.listdir():
    if 'historical_fx_rates' in f and not f.endswith('.zip'):
        print(f)
      
        for file in os.listdir(path=f):
            if file.endswith('.csv'):
                results = {'pair':[], 'Timestamp':[], 'VWAP':[],
            'Liquidity':[], 'volatility':[], 'max':[], 'min':[], 'FD':[]}
                print(file)
                df = pd.read_csv(f'{f}/{file}')
                df = df[(df['datetime'] >= start_date) & (df['datetime'] <= end_date)]
                #df=df.iloc[:10, :]
                df['hour']= pd.to_datetime(df['datetime']).dt.floor('H')
                #replace the NaN with average value of the column
                check_na_col = ['v', 'vw', 'o', 'c', 'h', 'l', 't', 'n',]
                for col in check_na_col:
                    if df[col].isnull().sum() > 0:
                        df[col] = df[col].fillna(df[col].mean())
                print('Data loaded')
                #calculate the vwap
                #df['VWAP'] = df['vw'] * df['v']

                #calculate features for each 6-hour interval
                for st in tqdm(pd.date_range(start_date, end_date, freq='6H')):
                    et = st + pd.Timedelta(hours=6)
                    df['datetime'] = pd.to_datetime(df['datetime'])
                    six_hour_delta = df[(df['datetime'] >= st) & (df['datetime'] < et)]
                    

                    if not six_hour_delta.empty:
                        #VWAP price for the current 6-hour interval
                        vwap_price = six_hour_delta['vw'].iloc[-1]
                        #Max, Min, and Mean VWAP price for the current 6-hour interval
                        max_price = six_hour_delta['h'].max()
                        min_price = six_hour_delta['l'].min()
                        mean_price = six_hour_delta['vw'].mean()
                        
                        #Total liquidity for the current 6-hour interval
                        liquidity = six_hour_delta['v'].mean()
                        #Volatility
                        vol = (max_price - min_price) / mean_price if mean_price else None
                        #kelner channel
                        kelner_upper = mean_price + 0.025 * vol
                        kelner_lower = mean_price - 0.025 * vol
                        #Count N, the #of times the price crosses the upper or lower kelner channel
                        crosses = ((six_hour_delta['vw'] > kelner_upper) | (six_hour_delta['vw'] < kelner_lower)).sum()
                        #FD
                        fd = crosses / (max_price - min_price)
                        #print(fd)
                        #Save the results
                        results['pair'] += [file]
                        results['Timestamp'] += [st]
                        results['VWAP'] += [vwap_price]
                        results['Liquidity'] += [liquidity]
                        results['volatility'] += [vol]
                        results['max'] += [max_price]
                        results['min'] += [min_price]
                        results['FD'] += [fd]
                results = pd.DataFrame(results)
                file_ = file.split('.')[0]
                output_path = f'/Users/apple/Desktop/Data Engineering/hw3/{file_}.csv'
                results.to_csv(output_path, index=False)
                

# output_path = '/Users/apple/Desktop/Data Engineering/hw3/results.csv'
# results_all.to_csv(output_path, index=False)



```

```python
#repeat the process and insert them into SQLite and MongoDb
#connect data to SQLite

import os
import pandas as pd
from pymongo import MongoClient
import sqlite3
from tqdm import tqdm


# SQLite connection details
database_path_q2 = '/Users/apple/Desktop/Data Engineering/hw3/hw3_step2.db'
conn_q2 = sqlite3.connect(database_path_q2)
cursor_q2 = conn_q2.cursor()

# MongoDB connection details
client_q2 = MongoClient('localhost', 27017)
db_q2 = client_q2['hw3_q2']

# Loop through all files in the current directory
for f in os.listdir():
    if 'historical_fx_rates' in f and not f.endswith('.zip'):
        print(f)
        
        for file in os.listdir(path=f):
            if file.endswith('.csv'):
                results = {'pair': [], 'Timestamp': [], 'VWAP': [],
                           'Liquidity': [], 'volatility': [], 'max': [],
                           'min': [], 'FD': []}
                print(file)
                df = pd.read_csv(f'{f}/{file}')
                df = df[(df['datetime'] >= start_date) & (df['datetime'] <= end_date)]
                df['hour'] = pd.to_datetime(df['datetime']).dt.floor('H')
                
                # Replace NaN with the average value of the column
                check_na_col = ['v', 'vw', 'o', 'c', 'h', 'l', 't', 'n']
                for col in check_na_col:
                    if df[col].isnull().sum() > 0:
                        df[col] = df[col].fillna(df[col].mean())
                print('Data loaded')

                # Calculate features for each 6-hour interval
                for st in tqdm(pd.date_range(start_date, end_date, freq='6H')):
                    et = st + pd.Timedelta(hours=6)
                    df['datetime'] = pd.to_datetime(df['datetime'])
                    six_hour_delta = df[(df['datetime'] >= st) & (df['datetime'] < et)]
                    
                    if not six_hour_delta.empty:
                        # VWAP price for the current 6-hour interval
                        vwap_price = six_hour_delta['vw'].iloc[-1]
                        # Max, Min, and Mean VWAP price for the current 6-hour interval
                        max_price = six_hour_delta['h'].max()
                        min_price = six_hour_delta['l'].min()
                        mean_price = six_hour_delta['vw'].mean()
                        # Total liquidity for the current 6-hour interval
                        liquidity = six_hour_delta['v'].mean()
                        # Volatility
                        vol = (max_price - min_price) / mean_price if mean_price else None
                        # Kelner channel
                        kelner_upper = mean_price + 0.025 * vol
                        kelner_lower = mean_price - 0.025 * vol
                        # Count N, the number of times the price crosses the upper or lower Kelner channel
                        crosses = ((six_hour_delta['vw'] > kelner_upper) | (six_hour_delta['vw'] < kelner_lower)).sum()
                        # FD
                        fd = crosses / (max_price - min_price)
                        
                        # Insert the results into SQLite
                        table_name = file.split('/')[-1].split('.')[0]
                        cursor_q2.execute(f'DROP TABLE IF EXISTS {table_name}')
                        cursor_q2.execute(f"""
                            CREATE TABLE {table_name} (
                                pair TEXT,
                                Timestamp TEXT,
                                VWAP REAL,
                                liquidity REAL,
                                volatility REAL,
                                max REAL,
                                min REAL,
                                FD REAL
                            )
                        """)
                        
                        cursor_q2.execute(f"""
                            INSERT INTO {table_name} (pair, Timestamp, VWAP, liquidity, volatility, max, min, FD)
                            VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                        """, (
                            file, st.strftime('%Y-%m-%d %H:%M:%S'), vwap_price, liquidity, vol, max_price, min_price, fd
                        ))
                        
                        # Insert the results into MongoDB
                        collection_name = file.split('.')[0]
                        collection = db_q2[collection_name]
                        collection.insert_one({
                            'pair': file,
                            'Timestamp': st,
                            'VWAP': vwap_price,
                            'Liquidity': liquidity,
                            'volatility': vol,
                            'max': max_price,
                            'min': min_price,
                            'FD': fd
                        })

# Commit and close the SQLite connection
conn_q2.commit()
cursor_q2.close()
conn_q2.close()

# Close the MongoDB connection
client_q2.close()

```
